```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pathlib import Path

DATA_DIR = Path("../data/raw")

file = list(DATA_DIR.glob("*.csv"))[0]

df = pd.read_csv(file)

df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
#Churn Distribution
df["Churned"].value_counts()
```

```python
df["Churned"].value_counts().plot(kind="bar")

plt.title("Churn Distribution")
plt.xlabel("Churn Status")
plt.ylabel("Number of Customers")

plt.show()
```

```python
#Engagement vs Churn
df.groupby("Churned")[[
    "Login_Frequency",
    "Session_Duration_Avg",
    "Pages_Per_Session",
    "Email_Open_Rate"
]].mean()
```

```python
#Purchase Behaviour vs Churn
df.groupby("Churned")[[
    "Total_Purchases",
    "Average_Order_Value",
    "Days_Since_Last_Purchase"
]].mean()
```

```python
#Correlation with Churn
corr = df.corr(numeric_only=True)

corr["Churned"].sort_values(ascending=False)
```

```python
#Country-Level Churn
df.groupby("Country")["Churned"].mean().sort_values(ascending=False)
```
